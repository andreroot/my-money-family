```python
from v0.resource.myprocessanalytics_v0 import process, analytics_percent, analytics_item_percent_diario, analytics_percent_2025

def get_fixo(ano):
    
    strsql = f"SELECT * FROM dev_domestico.custo_{ano}_excel "

    df = process(strsql, flag_fake=False)

    df = df[['tipo_custo','custo','valor_custo','dt_mes_base','dt_custo','tipo_custo_alt','classificacao_custo','area_custo']].drop_duplicates()

    return df
```
